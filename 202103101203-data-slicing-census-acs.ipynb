## Dataframe for Total Male Estimates 5 - 9 years

B01001_004E = Estimate!!Total!!Male!!5 to 9 years

```python
url4 = "https://api.census.gov/data/2017/acs/acs5?get=NAME,B01001_004E&for=block%20group:*&in=state:06%20county:073"

# San Diego County = 073
# in url B01001_004E = Estimate!!Total!!Male!!5 to 9 years

response = urlopen(url4)
data = loads(response.read())

data
;
```

```python
df4 = pd.DataFrame.from_dict(data)
df4
```

```python
df4.columns = df4.iloc[0]

#set first row as the column names in the dataframe

df4
```

```python
df4 = df4.iloc[1:]
df4

#delete row 0 after renaming columns 
```

# Geographic Unit Variables from Urban 

```python
from urllib.request import urlopen
from json import loads
url = "https://educationdata.urban.org/api/v1/schools/nhgis/census-2010/2016/?state_mailing=CA&fips=06&county_fips_geo=06073"
response = urlopen(url)
data = loads(response.read())

# county_fips_geo = 06073 (San Diego)
# years available: 1986 - 2016 
# year = 2016
```

```python
urban = pd.json_normalize(data, 'results')

urban
```
